```python
def  dicc_songs(give_list):
    artist_data = {}
    limit =30
    
    for i in give_list:
        results=sp.search(q=i,limit=30)

        for j in range(limit):
            artist_id = results['tracks']['items'][0]['artists'][0]['id']
            artist = results['tracks']['items'][0]['artists'][0]['name']
            song_name = results['tracks']['items'][j]['name']
            if artist_id not in artist_data:
                artist_data[artist_id] = [[song_name, artist]]
            else:
                duplicate_song = False
                for song in artist_data[artist_id]:
                    if song[0] == song_name:
                        duplicate_song = True
                        break
                if not duplicate_song:
                    artist_data[artist_id].append([song_name, artist])
    return artist_data
```

```python
song_name = search
def search_same_artist(song_name,dicttionary):
    # Buscar el artista de la canción
    artist_id = None
    for id, songs in dicttionary.items():
        for song in songs:
            if song[0] == song_name:
                artist_id = id
                break
        if artist_id:
            break
            
    # Recomendar una canción aleatoria del mismo artista
    if artist_id:
        artist_songs = dicttionary[artist_id]
        recommended_song = None
        other_songs = [song[0] for song in artist_songs if song[0] != song_name]
        if other_songs:
            recommended_song = random.choice(other_songs)
        if recommended_song:
            result = print(f"Si te gusta {song_name}, te podría gustar {recommended_song} de este mismo artista.")
        else:
            result = print("No se encontró otra canción de este mismo artista.")
    else:
        result = print("No se encontró la canción en el diccionario.")
    
    return result
```
