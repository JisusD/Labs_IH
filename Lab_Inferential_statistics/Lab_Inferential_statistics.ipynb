# Lab Inferential statistics

## Instructions
It is assumed that the mean systolic blood pressure is μ = 120 mm Hg. In the Honolulu Heart Study, a sample of n = 100 people had an average systolic blood pressure of 130.1 mm Hg with a standard deviation of 21.21 mm Hg. Is the group significantly different (with respect to systolic blood pressure!) from the regular population?

- Set up the hypothesis test.

- Write down all the steps followed for setting up the test.

Para responder a esta pregunta, podemos realizar una prueba de hipótesis utilizando la distribución t de Student. La hipótesis nula (H0) es que la presión arterial sistólica media de la población es igual a 120 mm Hg, y la hipótesis alternativa (HA) es que la presión arterial sistólica media de la población es diferente de 120 mm Hg.

Podemos plantear la hipótesis nula y alternativa de la siguiente manera:

H0: μ = 120
HA: μ ≠ 120

El nivel de significancia que usaremos para esta prueba será de α = 0.05.

Para realizar la prueba de hipótesis, primero calculamos el estadístico de prueba t utilizando la fórmula:

t = (x̄ - μ) / (s / √n)

Donde x̄ es la media muestral, μ es la media poblacional hipotética, s es la desviación estándar muestral y n es el tamaño de la muestra.

Reemplazando los valores conocidos en la fórmula, obtenemos:

t = (130.1 - 120) / (21.21 / √100) = 4.76

Luego, podemos buscar el valor crítico de t para un nivel de significancia de α = 0.05 y 98 grados de libertad (n-1). Buscando en una tabla de distribución t, encontramos que el valor crítico de t es aproximadamente ±1,984.

Como el valor absoluto del estadístico de prueba (4.76) es mayor que el valor crítico de t (1.984), podemos rechazar la hipótesis nula y concluir que hay evidencia suficiente para afirmar que la presión arterial sistólica media de la población no es igual a 120 mm Hg. En otras palabras, el grupo es significativamente diferente de la población normal con respecto a la presión arterial sistólica.

- Calculate the test statistic by hand and also code it in Python. It should be 4.76190. We will take a look at how to make decisions based on this calculated value.

```python
import numpy as np

# Datos de la muestra
n = 100
xbar = 130.1
s = 21.21

# Hipótesis nula
mu = 120

# Usando la formula para T-score 
top_eq = xbar - mu
bottom_eq = s / (np.sqrt(n))

# Nivel de significancia
alpha = 0.05

# Imprimir resultados
t = top_eq / bottom_eq
print(t)
```
